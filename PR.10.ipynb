```python
import os
import subprocess
import psutil

class case_1:
    def __init__(self):
        while True:
            print("\nFile System Operations")
            print("1. Display current directory")
            print("2. List files and directories")
            print("3. Change directory")
            print("4. Create a directory")
            print("5. Remove a directory")
            print("6. Create file")
            print("7. Delete a file")
            print("8. Back to Main Menu")
            print("===================")
            self.choice = int(input("Enter your choice: "))
            print()

            match self.choice:
                case 1:
                    print("Current directory:", os.getcwd())

                case 2:
                    print("Files and directories in current directory:", os.listdir())

                case 3:
                    self.path = input("Enter the path to change to: ")
                    try:
                        os.chdir(self.path)
                        print("Directory changed to:", self.path)
                        print("Current directory:", os.getcwd())
                    except FileNotFoundError:
                        print("Directory not found")

                case 4:
                    self.name = input("Enter the name of the directory to create: ")
                    try:
                        os.mkdir(self.name)
                        print("Directory", self.name , "created successfully.")
                        print("Current directory:", os.getcwd())
                    except FileExistsError:
                        print("Directory already exists")

                case 5:
                    self.name = input("Enter the name of the directory to remove: ")
                    try:
                        os.removedirs(self.name)
                        print("Directory removed:", self.name)
                        print("Current directory:", os.getcwd())
                    except FileNotFoundError:
                        print("Directory not found")

                case 6:
                    self.name = input("Enter the name of the file to create: ")
                    try:
                        open(self.name, 'w').close()
                        print("File created:", self.name)
                        print("Current directory:", os.getcwd())
                    except FileExistsError:
                        print("File already exists")

                case 7:
                    self.name = input("Enter the name of the file to delete: ")
                    try:
                        os.remove(self.name)
                        print("File deleted:", self.name)
                        print("Current directory:", os.getcwd())
                    except FileNotFoundError:
                        print("File not found")

                case 8:
                    break

                case _:
                    print("Invalid option. Please choose a valid option.")

class case_2:
    def __init__(self):
        while True:
            print("\nExecute Terminal Commands")
            print("1. Run a predefined command")
            print("2. Run a custom command")
            print("3. Back to Main Menu")
            print("===================")
            self.choice = int(input("Enter your choice: "))
            print()

            match self.choice:
                case 1:
                    print("Choose a command:")
                    print("1. pwd")
                    print("2. ls")
                    print("3. df -h")
                    print("===================")
                    self.ch = int(input("Enter your choice: "))
                    if self.ch == 1:
                        print("Current directory:",os.getcwd())
                    elif self.ch == 2:
                        print("Files in the current directory:", os.listdir())
                    elif self.ch == 3:
                        self.disk_usage = psutil.disk_usage('/')
                        print("Disk usage:")
                        print(f"Total: {self.disk_usage.total / (1024.0 ** 3):.2f} GB")
                        print(f"Used: {self.disk_usage.used / (1024.0 ** 3):.2f} GB")
                        print(f"Free: {self.disk_usage.free / (1024.0 ** 3):.2f} GB")
                        print(f"Percentage: {self.disk_usage.percent}%")
                    else:
                        print("Invalid choice")

                case 2:
                    self.command = input("Enter the command to execute: ")
                    try:
                        self.output = subprocess.check_output(self.command, shell=True)
                        print(self.output.decode('utf-8'))
                    except subprocess.CalledProcessError as e:
                        print("Error:", e)

                case 3:
                    break

                case _:
                    print("Invalid option. Please choose a valid option.")

class case_3:
    def __init__(self):
        while True:
            print("\nFile Analysis with Higher-Order Functions")
            print("1. Sort files by name")
            print("2. Filter files by extension")
            print("3. Calculate total size of files in a directory")
            print("4. Back to Main Menu")
            print("===================")
            self.choice = int(input("Enter your choice: "))
            print()

            match self.choice:
                case 1:
                    self.files = os.listdir()
                    self.sorted_files = sorted(self.files)
                    print("Sorted files:", self.sorted_files)

                case 2:
                    self.file_choice = input("Enter the file extension (e.g., .py, .txt): ")
                    self.found_file = False
                    self.found_file_extension = False

                    for self.file in os.listdir():
                        self.found_file = True
                        if self.file.endswith(self.file_choice):
                            self.found_file_extension = True
                            print(self.file)

                    if not self.found_file:
                        print("No files found in current directory.")
                    elif not self.found_file_extension:
                        print("No files with the specified extension found in current directory.")
                    
                case 3:
                    self.files = os.listdir()
                    self.total_size = sum(os.path.getsize(self.file) for self.file in self.files)
                    print(f"Total size : {self.total_size} bytes")

                case 4:
                    break

                case _:
                    print("Invalid choice")
                
# main program

print("===================")
print("Execter - Command Line Utility and File System Manager")
print("===================")

while True:
    print("\nChoose an option:")
    print("1. File System Operations")
    print("2. Execute Terminal Commands")
    print("3. File Analysis with Higher-Order Functions")
    print("4. Exit")
    print("===================")
    option = int(input("Enter your option: "))

    match option:
        case 1:
            C1 = case_1()
               
        case 2:
            C2 = case_2()

        case 3:
            C3 = case_3()

        case 4:
            permission = input("Do you really want to exit? (yes/no): ")
            if permission.lower() == "yes":
                print("\nThank you for using the Executer Manager. Goodbye! Command Line Utility and File System.")
                break
            elif permission.lower() == "no":
                print("\nYou have chosen to cancel the exit. You will be returned to the main menu.")
                continue
            else:
                print("\nInvalid input. Please enter 'yes' or 'no' to confirm your choice.")

        case _:
            print("Invalid option. Please choose a valid option.")
```
