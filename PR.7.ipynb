```python
import datetime
import os

class JournalManager:

    def init(self):
        try:
            with open("Journal.txt", "r") as self.f:
                pass
        except FileNotFoundError:
            self.f = open("Journal.txt" , "x")
        
        self.f.close()

    def adding(self):
        self.f = open("Journal.txt" , "a")
        self.entry = input("Enter your journal entry : ")
        self.current_time = datetime.datetime.now()
        self.a = self.current_time.strftime("[%Y-%m-%d %H:%M:%S]")

        if not self.entry:
            print("Please enter a valid entry")
        else:
            self.f.write(f"{self.a} : {self.entry}\n")
            self.f.close()
            print("Entry added successfully")

    def reading(self):
        try:
            with open("Journal.txt" , "r") as self.f: 
                self.content = self.f.read()
                if not self.content:
                    print("No journal entries found. Start by adding a new entry!")
                else:
                    print(self.content)
                    self.f.close()
        except FileNotFoundError:
            print("Error: The Journal file dose not exist. Please create file first.")

    def Searching(self,search):
        self.search = search
        self.f = open("Journal.txt" , "r")
        self.content = self.f.readlines()
        self.found = False
        for line in self.content:
            if self.search in line:
                self.found = True
                print("Matching Entries:\n------------------------")
                print(line.strip())
        self.f.close()

        if not self.found:
            print(f"No entries were found for the keyword: {self.search}")

    def deleting(self):
        self.option = input("Are you sure you want to delete file? (Yes/ No) : ")
        if self.option.lower() == "yes":
            if os.path.exists("Journal.txt"): 
                os.remove("Journal.txt")
                print("File deleted!")
            else:
                print("File does not exist!")
        else:
            print("Deletion cancelled")
            
# Main program

print("Welcome to Personal Journal Manager!")
JM = JournalManager()

while True:
    print("\nPlease select an option:")
    print("1. Add a New Entry")
    print("2. View All Entries")
    print("3. Search for an Entry")
    print("4. Delete All Entries")
    print("5. Exit")
    choice = int(input("Enter your choice : "))
    print()

    match choice:
        case 1:
            # Add a New Entry
            JM.adding()

        case 2:
            # View All Entries
            JM.reading()

        case 3:
            # Search for an Entry
            search = input("Enter a keyword or date to search : ")
            JM.Searching(search)

        case 4:
            # Delete All Entries
            JM.deleting()

        case 5:
            # Exit
            print("Thank you for using Personal Journal Manager.\nGoodbye!")
            break

        case _:
            print("Invalid choice. Please select a valid option from menu.")
```
