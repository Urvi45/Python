```python
print("Wellcome to the Pattern Generator and Number Analyzer!")
choice = 1
while choice<3:
  print("\nSelect an option:\n1. Generate a Pattern\n2. Analyze a Range of Numbers\n3. Exit")
  choice = int(input("Enter your choice:"))
  match(choice):
    case 1:
      print("\nChoose a pattern type:\n1. Right-angled Triangle\n2. Pyramid\n3. Left-angled Triangle")
      choose = int(input("Enter your choice:"))
      n = int(input("Enter the number of rows for the pattern:"))
      if choose == 1:
        for i in range(1,n+1):
          for j in range(1,i+1):
            print(" ",end = "*")
          print()
      elif choose == 2:
        for i in range(1 ,n+1):
          for j in range(n-i):
            print(" ",end = " ")
          for k in range(1, 2*i):
            print("*" , end = " ")
          print()
      elif choose == 3:
        for i in range(1,n+1):
          for j in range(1,n-i+1):
            print(" " , end = " ")
          for j in range(1, i+1):
            print("*",end = " ")
          print()
      else:
        print("invalid syntax")
    case 2:
      start = int(input("\nEnter the start of the range:"))
      end = int(input("Enter the end of the range:"))
      for i in range(start , end+1):
        if i % 2 ==0:
          print("Number",i,"is Even")
        else:
          print("Number",i,"is Odd")
      print("Sum of all numbers from",start,"to",end,"is:",sum(range(start , end+1)))
    case 3:
      print("Exiting the program. Goodbye!")
```
