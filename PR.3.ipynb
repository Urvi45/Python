```python
print("Welcome to the Student Data Organizer!")

option = 1
student_records = []
sub_offered = set()

while(option<6):

    print("\nSelect an option:")
    print("1. Add Student")
    print("2. Display All Students")
    print("3. Update Student Information")
    print("4. Delete Student")
    print("5. Display Subjects Offered")
    print("6. Exit")

    option = int(input("Enter your choice : "))
    print()

    match option:

        case 1:
            n = int(input("Enter number of student : "))

            for i in range(n):
                print(f"\nEnter {i+1} student details : ")
                Id = int(input("Enter student ID : "))
                name = input("Enter student name : ")
                age = int(input("Enter student age : "))
                grade = input("Enter student grade : ")
                bob = input("Enter student date of Birth (YYYY-MM-DD) : ")
                subject = input("Subjects (comma-separated) : ").split(",")

                student ={
                    "Id" : Id,
                    "Name" : name,
                    "Age" : age,
                    "Grade" : grade,
                    "BOB" : bob,
                    "Subject" :[sub.strip() for sub in subject]
                }

                student_records.append(student)

                for sub in student["Subject"]:
                    sub_offered.add(sub)

        case 2:

            for student in student_records:
                print(f"Student ID : { student['Id']} | Name : {student['Name']} | Age : {student['Age']} | Grade : {student['Grade']} | Subjects : {', '.join(student['Subject'])} | Date of Birth : {student['BOB']}")

        case 3:
            update = int(input("Enter student ID to update: "))
            s_found = False

            for student in student_records:
                if student['Id'] == update:
                    s_found = True
                    print("Which information would you like to update?")
                    print("1. Age")
                    print("2. Subjects")
                    update_option = int(input("Enter choice: "))

                    if update_option == 1:
                        new_age = int(input("Enter new age: "))
                        student['Age'] = new_age
                        print(f"Student ID {update} age updated to {new_age}.")
                    elif update_option == 2:
                        subject = input("Enter new subjects (comma-separated): ").split(",")
                        student['Subject'] = [sub.strip() for sub in subject]

                        print(f"Student ID {update} subjects updated.")
                    else:
                        print("Invalid option.")
                    break

            if not s_found:
                print(f"Student ID {update} not found.")

        case 4:
            Id = int(input("Enter student ID to delete: "))
            s_found = False
            for i , student in enumerate(student_records):
                if student['Id'] == Id:
                    s_found = True
                    del student_records[i]
                    print(f"Student with ID {Id} has been deleted.")
                    break

            if not s_found:
                print("Student ID not found.")

        case 5:

            print("Unique Subjects Offered:")
            for sub in sub_offered:
                print(sub)

        case 6:
          print("Thanks for using the Student Data Oraganizer.")
```
