```python
import requests
import html5lib 
from bs4 import BeautifulSoup 

print("=================================")
print("Welcome to Scraper & Bot")
print("A Web Scraping and Automation Utility")
print("=================================")

while True:
    print("\nChoose an option:")
    print("1. Web Scraping")
    print("2. Web Automation")
    print("3. Combined Tasks (Scraping + Automation)")
    print("4. Exit")
    print("=================================\n")
    choice = int(input("Enter your choice:"))

    match choice:
        case 1:
            print("\nWeb Scraping Options:")
            print("1. Scrape Blog Data (Titles and Links)")
            print("2. Scrape E-commerce Products (Name, Price, Rating)")
            print("3. Scrape Weather Data (Temperature, Conditions)")
            print("4. Back to Main Menu")
            print("=================================\n")
            scrape_choice = int(input("Enter your choice: "))

            match scrape_choice:
                case 1:
                    Blog_Data = input("Enter the URL to scrape :")
                    response = requests.get(Blog_Data)
                    soup = BeautifulSoup(response.text, 'html5lib')
                    print("title : " , soup.title.string)

                    links = soup.find_all('a')
                    len_links = len(links)

                    print("Total link in this URl : ", len_links)

                    if len_links > 0:
                        link = links[1:11]
                        for i in link:
                            print(i.get('href'))
                        

                case 2:
                    url = "https://www.amazon.in/s?bbn=1375424031&rh=n%3A1375424031%2Cp_89%3AHP&pf_rd_i=1375424031&pf_rd_m=A1VBAL9TL5WCBF&pf_rd_p=3731bbfb-e3cd-40ae-a3dd-8d5a38232c06&pf_rd_r=ED6EBQ1J9108QYFM34XM&pf_rd_s=merchandised-search-6&ref=AF_WIN_bub_w_cml_t_1"
                    response = requests.get(url)

                    soup = BeautifulSoup(response.text, 'html5lib')
                    products = soup.find_all('div', class_='a-section a-spacing-base')
                    rate = soup.find_all('span', class_='a-icon-alt')
                    price = soup.find_all('span', class_='a-price-whole')

                    len_of_item = len(products)
                    if len_of_item > 0:
                        length1 = products[0:5]
                        length2 = price[0:5]
                        length3 = rate[0:5]
                        for i in range(len(length1)):
                            print(f"Product Name: {length1[i].text}")
                            print(f"Product Price: {length2[i].text}")
                            print(f"Product Rating: {length3[i].text}")
                        
                case 3:
                    print("Scraping Weather Data...")
                    url = "https://www.accuweather.com/en/in/surat/202441/current-weather/202441"

                    response = requests.get(url, headers={'User-Agent': 'Mozilla/5.0'})
                    soup = BeautifulSoup(response.text, 'html.parser')


                    temperature_elements = soup.find_all('div', class_="display-temp")
                    if temperature_elements:
                        temperature = temperature_elements[0].text.strip()
                        print(f"Temperature in surat : {temperature}")
                        print("Weather Data Scraped Successfully!")
                    else:
                        print("Temperature not found. Structure of the page might have changed.")

                case 4:
                    break
                
                case _:
                    print("Invalid choice. Please try again.")

        case 2:
            print("\nWeb Automation Options:")
            print("1. Login to a Website")
            print("2. Fill and Submit a Form")
            print("3. Search and Capture Google Results")
            print("4. Back to Main Menu")
            print("=================================\n")
            Automation_choice = int(input("Enter your choice: "))

            match Automation_choice:
                case 1:
                    print("Login to a Website...")
                    url = "https://www.example.com/login"
                    username = "your_username"
                    password = "your_password"
                    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
                    driver.get(url)
                    driver.find_element(By.NAME, "username").send_keys(username)
                    driver.find_element(By.NAME, "password").send_keys(password)
                    driver.find_element(By.NAME, "login").click()
                    print("Login Successful!")
                    

                case 2:
                    pass

                case 3:
                    pass

                case 4:
                    break
                case _:
                    print("Invalid choice. Please try again.")

        case 3:
            print("\nCombined Tasks Options:")
            print("1. Scrape Search Results and Visit Links")
            print("2. Scrape Products From an E-Commerce Website")
            print("3. Back to Main Menu")
            print("=================================\n")
            Combined_choice = int(input("Enter your choice: "))

            match Combined_choice:
                case 1:
                    pass

                case 2:
                    pass

                case 3:
                    break
                case _:
                    print("Invalid choice. Please try again.")

        case 4:
            # exit = input("Do you really want to exit? (yes/no):").lower()
            # if exit == "yes":
            #     print("Thank you for using Scraper & Bot! Goodbye!")
            #     break
            # else:
            #     continue
            print("Thank you for using Scraper & Bot! Goodbye!")
            break





```
