```python
class Employee:
    name = None
    age = 0
    salary = 0
    id = None

    def getter(self):
        self.name = input("Enter your name : ")
        self.age = int(input("Enter your age : "))
        self.id = input("Enter your id : ")
        self.salary = int(input("Enter your salary : "))

    def setter(self):
        print(f"Name : {self.name}")
        print(f"Age : {self.age}")
        print(f"ID : {self.id}")
        print(f"Salary : {self.salary}")

    def __eq__(self, other):
        return self.salary == other.salary

    def __lt__(self, other):
        return self.salary < other.salary

    def __gt__(self, other):
        return self.salary > other.salary


class Manager(Employee):
    department = None

    def getter(self):
        super().getter()
        self.department = input("Enter your department : ")

    def setter(self):
        super().setter()
        print(f"Department : {self.department}")

class Developer(Employee):
    program_language = None

    def getter(self):
        super().getter()
        self.program_language = input("Enter your programming language : ")

    def setter(self):
        super().setter()
        print(f"Programming language : {self.program_language}")


# main program

print("--- Python OOP Project: Employee Management System ---")

choice = 1

employee = []

while choice < 6:

    print("\nChoose an operation:")
    print("1. Create a Employee")
    print("2. Create an Manager")
    print("3. Create a Developer")
    print("4. Show Details")
    print("5. Compare Salaries")
    print("6. Exit\n")

    choice = int(input("Enter your choice : "))

    match choice:
        case 1:
            emp = Employee()
            emp.getter()
            employee.append(emp)
            print(f"Employee created with name: {emp.name}, age: {emp.age}, ID: {emp.id}, and salary: ${emp.salary}.")
            print("\n --- Choose another operation ---")

        case 2:
            manager = Manager()
            manager.getter()
            employee.append(manager)
            print(f"Employee created with name: {manager.name}, age: {manager.age}, ID: {manager.id}, salary: ${manager.salary}, and department: {manager.department}.")
            print("\n --- Choose another operation ---")

        case 3:
            developer = Developer()
            developer.getter()
            employee.append(developer)
            print(f"Employee created with name: {developer.name}, age: {developer.age}, ID: {developer.id}, salary: ${developer.salary}, and programming language: {developer.program_language}.")
            print("\n --- Choose another operation ---")

        case 4:
            print("Choose details to show:")
            print("1. Employee")
            print("2. Manager")
            print("3. Developer")

            choose = int(input("Enter your choice : "))

            if choose == 1:
                print("\nEmployee Details:")
                for emp in employee:
                    if isinstance(emp, Manager):
                        pass
                    elif isinstance(emp, Developer):
                        pass
                    else:
                        emp.setter()
            elif choose == 2:
                print("\nManager Details:")
                for emp in employee:   
                    if isinstance(emp, Manager):
                        manager.setter()
            elif choose == 3:
                print("\nDeveloper Details:")
                for emp in employee:    
                    if isinstance(emp, Developer):
                        developer.setter()
            else:
                print("Invalid choice! Please choose a valid option.")
            print("\n --- Choose another operation ---")

        case 5:
            print("Salary Comparison:")
            id1 = input("Enter the first employee's ID : ")
            id2 = input("Enter the second employee's ID : ")

            emp1 = next((emp for emp in employee if emp.id == id1), None)
            emp2 = next((emp for emp in employee if emp.id == id2), None)

            if emp1 and emp2:
                if emp1 == emp2:
                    print("Both employees have the same salary.")
                elif emp1 < emp2:
                    print(f"Employee {emp1.name} has a lower salary than Employee {emp2.name}.")
                else:
                    print(f"Employee {emp1.name} has a higher salary than Employee {emp2.name}.")
            else:
                print("One or both employees not found.")
            print("\n --- Choose another operation ---")
            
        case 6:
            print("Exiting the system. All resources have been freed.")
            print("Goodbye!")

        case _:
            print("Invalid choice! Please choose a valid option.")
```
