```python
class MyException(Exception):
    def __init__(self, message):
        self.message = message

    def display_message(self):
        print(f"Error: {self.message}")

class InsufficientFundsError(Exception):
    def __init__(self, message):
        self.message = message

    def display_message(self):
        print(f"Error: {self.message}")

# main program

print("Welcome to the Robust Banking System!")

choice = 1

while choice < 5:
    print("\nPlease select an option:")
    print("1. Create Account")
    print("2. Deposit Funds")
    print("3. Withdraw Funds")
    print("4. Check Balance")
    print("5. Exit")

    choice = int(input("Enter your choice : "))
    try:
        assert choice <= 5
    except:
        print("Invalid choice. Please choose a valid option.")

    match choice:
        case 1:
            name = input("Enter your name : ")
            amount = int(input("Enter initial deposit amount : "))
            try:
                assert amount > 0
            except:
                print("Amount must be greater than 0.")
            else:
                print("Account created successfully!")

        case 2:
            deposit_amount = int(input("Enter deposit amount : "))
            
            try:
                if deposit_amount < 0:
                    raise MyException("Deposit amount must be positive.")
                else:
                    amount += deposit_amount
                    print(f"Deposit successful! Your new balance is : {amount}")
            except MyException as d:
                    d.display_message()
            
        case 3:
            withdraw_amount = int(input("Enter withdraw amount : "))

            try:
                if withdraw_amount > amount:
                    raise InsufficientFundsError(f"Insufficient funds. Your current balance is: {amount}")
                else:
                    amount -= withdraw_amount
            except InsufficientFundsError as w:
                    w.display_message()
            else:
                print(f"Withdraw successful! Your new balance is : {amount}")

        case 4:
            print(f"Your current balance is: {amount}")

        case 5:
            print("Thank you for using Robust Banking System! Goodbye!")
```
